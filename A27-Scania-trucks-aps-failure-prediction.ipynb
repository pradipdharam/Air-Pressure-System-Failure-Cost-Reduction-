## Scania Trucks APS Failure Predictions

Purpuse is to minimize maintenance costs of the air pressure system (APS) of Scania trucks.
Therefore, failures should be predicted before they occur. Falsely predicting a
failure has a cost of 10, missing a failure a cost of 500. This leads to the need of
cost minimization. The datasets' positive class consists of component failures for a specific component of the APS system. The negative class consists of trucks with failures for components not related to the APS.

### Challenge metric 

 Cost-metric of miss-classification: 

 Predicted class | True class | 
 | pos | neg | 
  
 pos | - | Cost_1 | 
 
 neg | Cost_2 | - | 
 
 Cost_1 = 10 and cost_2 = 500 

 The total cost of a prediction model the sum of 'Cost_1' 
 multiplied by the number of Instances with type 1 failure 
 and 'Cost_2' with the number of instances with type 2 failure, 
 resulting in a 'Total_cost'. 

 In this case Cost_1 refers to the cost that an unnessecary 
 check needs to be done by an mechanic at an workshop, while 
 Cost_2 refer to the cost of missing a faulty truck, 
 which may cause a breakdown. 

 Total_cost = Cost_1\*No_Instances + Cost_2\*No_Instances.

### Machine Learning Problem Statement:
Total_cost = Cost_1\*No_Instances + Cost_2\*No_Instances.

*  Total_cost = 10\*No_Instances + 500\*No_Instances it should be like this
*  Confusion Matrix:
* -----------------Actual	
* -----------------Yes	No
* Predicted	Yes	TP	FP
* -------------No	FN	TN

Here;
* Total_cost = Cost_1\*No_Instances + Cost_2\*No_Instances.
* As Cost_1 = 10 and cost_2 = 500 are explicitly given.
* Below is my understanding and I have rewritten the formula.
* Total_cost = cost_1 \* false positives  + cost_2 \* false negatives
* Total_cost = 10 \* false positives  + 500 \* false negatives 

And my goal is to have loss i.e. total_cost (its score in your email) less than < 9950.

## Solution:

### Importing required libraries

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.metrics import make_scorer
pd.options.display.max_columns = None
import seaborn as sns
import numpy as np
import matplotlib
matplotlib.use(u'nbAgg')
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
```

### Reading source files

```python
df_train = pd.read_csv("../input/aps_failure_training_set.csv")
df_test = pd.read_csv("../input/aps_failure_test_set.csv")
print("df_train.shape: ", df_train.shape)
print("df_test.shape: ", df_test.shape)
```

#### Mean imputations for respective features

```python
df_train['class'] = df_train['class'].replace({'neg': 0, 'pos': 1})
df_test['class']  = df_test['class'].replace({'neg': 0, 'pos': 1})

df_train.replace('na', np.NaN, inplace = True)
df_test.replace('na', np.NaN, inplace = True)

df_train = df_train.apply(pd.to_numeric)
df_test  = df_test.apply(pd.to_numeric)
```

#### Removing outliers

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df_train.describe(percentiles=[.1,.15,.2,.25,.3,.35,.4,.45, .5, .75,.8,.9,.95,.98,.99])
```

```python
df_train = df_train[df_train.ac_000 < 2130706432]
df_test = df_test[df_test.ac_000 < 2130706432]
print("df_train.shape: ", df_train.shape)
print("df_test.shape: ", df_test.shape)

df_train = df_train[df_train.ad_000 < 4000]
df_test = df_test[df_test.ad_000 < 4000]
print("df_train.shape: ", df_train.shape)
print("df_test.shape: ", df_test.shape)
```

```python
df_train.describe(percentiles=[.1,.15,.2,.25,.3,.35,.4,.45, .5, .75,.8,.9,.95,.98,.99])
```

#### train and test set creation, Imputation and Standardozation

```python
y_train = df_train['class'].values
x_train = df_train.drop(['class'], axis=1).values
y_test  = df_test['class'].values
x_test  = df_test.drop(['class'], axis=1).values
```

```python
imp = SimpleImputer(missing_values=np.NaN, strategy='mean')
x_train = imp.fit_transform(x_train)
x_test = imp.transform(x_test)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
#print(scaler.mean_)
x_train = scaler.transform(x_train)
x_test  = scaler.transform(x_test)
```

### Defining loss/cost function and scorer

```python
def total_cost_metric(y_test, y_pred):
    '''
    Total_cost = Cost_1*No_Instances + Cost_2*No_Instances.
    As Cost_1 = 10 and cost_2 = 500 are explicitly given.
    Below is my understanding and I have rewritten the formula.
    Total_cost = cost_1 * false positives  + cost_2 * false negatives
    '''
    #print(confusion_matrix(y_test, y_pred))
    tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()  
    return fp * 10 + fn * 500

# Make scorer and define that higher scores are better
total_cost_score = make_scorer(total_cost_metric, greater_is_better=False)
# Apply custom scorer to ridge regression
#total_cost_score(sig_clf, x_test, y_test)
```

### Model tuning

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV

x_cfl=XGBClassifier()
prams={
    #'learning_rate':[0.01, 0.05, 0.15, 0.2],
     'n_estimators':[10, 20, 50, 100, 150, 200, 250, 300, 350, 400]
     #'max_depth':[3,5,10]
     #'colsample_bytree':[0.5,1] #:[0.1,0.3,0.5,1]
     #'subsample':[0.1,0.3,0.5,1]   
}
random_cfl1=RandomizedSearchCV(x_cfl,param_distributions=prams, scoring =total_cost_score, \
                               verbose=10,n_jobs=-1)
%time random_cfl1.fit(x_train,y_train)
```

```python
n_estimators = random_cfl1.best_params_['n_estimators']
print("best n_estimators is ", n_estimators)

x_cfl=XGBClassifier(n_estimators=n_estimators, n_jobs=-1)
x_cfl.fit(x_train,y_train)

y_pred = x_cfl.predict(x_test)
total_cost_ = total_cost_metric(y_test, y_pred)
print("Optimal total cost is ", total_cost_)
```
